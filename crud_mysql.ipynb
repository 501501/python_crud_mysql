```python
from tkinter import * 
import tkinter.messagebox as MessageBox
import tkinter.ttk
```

```python
import pymysql
con=pymysql.connect(
    host="localhost", 
    user="root",
    password="",
    database='mydb'
)
```

```python
#창 만들기
root=Tk()
root.geometry("1300x300") #창 크기 지정
root.title("회원 정보 관리 시스템")

name=Label(root,text="이름") #라벨 지정
name.place(x=20,y=30) #위치 설정

birthday=Label(root,text=＂생일＂)
birthday.place(x=20,y=60)

tel=Label(root,text=＂전화번호＂)
tel.place(x=20,y=90)

e_name=Entry() #입력 받는 칸
e_name.place(x=150, y=30)

e_birthday=Entry()
e_birthday.place(x=150, y=60)

e_tel=Entry()
e_tel.place(x=150, y=90)

lbl=tkinter.Label(root,text=＂information＂)
lbl.pack()

#treeview를 이용하여 테이블 만들기
treeview=tkinter.ttk.Treeview(root,columns=("one,two,three","four"))
treeview.pack()

#열 지정
treeview["columns"]=("one","two","three","four")
treeview.column("#0",width=100,anchor="center")
treeview.column("#1",width=100,anchor="center")
treeview.column("#2",width=100,anchor="center")
treeview.column("#3",width=100,anchor="center")

#열 이름
treeview.heading("#0",text="no",anchor="center")
treeview.heading("#1",text="name",anchor="center")
treeview.heading("#2",text="birthday",anchor="center")
treeview.heading("#3",text="tel",anchor="center")

#삽입 함수
def insert():
    global treeview #전역변수
    Insert=＂Insert into student(name,birthday,tel) values (%s,%s,%s)“ #sql문 사용하여 삽입
    name=e_name.get() #입력한 이름 받아옴
    birthday=e_birthday.get() #입력한 생일 받아옴
    tel=e_tel.get() #입력한 번호 받아옴
    
    if(name!="" and birthday!="" and tel!=""): #이름, 생일, 번호 칸이 다 공백이 아닌 경우
        Value=(name,birthday,tel)
        cursor=con.cursor()
        cursor.execute(Insert,Value) #sql문 실행
        con.commit()
        MessageBox.showinfo(＂Insert Status＂,＂Inserted Successfully＂) #실행이 정상적으로 되었음을 창으로 알림
        
        treeview.destroy() #테이블 전체 삭제 후 
        #treeview.delete를 사용할 경우 노드가 삭제되지 않아 destro로 전체를 삭제한 후 표를 다시 생성하였음
        treeview=tkinter.ttk.Treeview(root,columns=("one,two,three","four"))
        treeview.pack()
        treeview["columns"]=("one","two","three","four")
        treeview.column("#0",width=100,anchor="center")
        treeview.column("#1",width=100,anchor="center")
        treeview.column("#2",width=100,anchor="center")
        treeview.column("#3",width=100,anchor="center")
        treeview.heading("#0",text="no",anchor="center")
        treeview.heading("#1",text="name",anchor="center")
        treeview.heading("#2",text="birthday",anchor="center")
        treeview.heading("#3",text="tel",anchor="center")
    
        ShowRecord() #전체를 조회하는 함수
        e_name.delete(0,＇end＇)
        e_birthday.delete(0,＇end＇)
        e_tel.delete(0,＇end＇) 
        
    else:
        MessageBox.showinfo(＂Insert Status＂,＂Some fields left blank＂) #입력칸이 다 채워지지 않은 경우 실행되지 않고 알림 창만 실행 

#전체 조회 함수
def ShowRecord():
    cursor=con.cursor()
    cursor.execute("select * from student") #sql문 실행
    rows=cursor.fetchall() #레코드를 배열 형식으로 저장
    cursor.close()
    for i in range(len(rows)): #배열을 treeview.insert를 이용해 표에 삽입
        treeview.insert("","end",text=i+1,values=rows[i],iid=str(i)+"번")

#검색 함수
def get():
    global treeview #전역함수
    if(e_name.get()==＂＂): #검색할 이름을 입력하지 않은 경우
        MessageBox.showinfo(＂Get Status＂,＂name is compolsary for get＂) #실패를 알리는 알림 창
    
    else: #검색할 이름을 입력한 경우
        cursor=con.cursor()
        val="%"+e_name.get()+"%“ #검색할 이름을 포함한 문자열 (%문자열%)
        cursor.execute(＂SELECT * FROM student WHERE name LIKE ＇%s＇＂ %(val)) #sql문
        rows=cursor.fetchall()
        cursor.close()
        MessageBox.showinfo(＂Get Status＂,＂Got Successfully＂) #검색 성공을 알리는 알림 창
        #삽입 함수에서와 같은 과정 (전체 테이블 삭제 후 다시 생성)
        treeview.destroy()
        treeview=tkinter.ttk.Treeview(root,columns=(＂one,two,three＂,＂four＂))
        treeview.pack()
        treeview[＂columns＂]=(＂one＂,＂two＂,＂three＂,＂four＂)
        treeview.column(＂#0＂,width=100,anchor=＂center＂)
        treeview.column(＂#1＂,width=100,anchor=＂center＂)
        treeview.column(＂#2＂,width=100,anchor=＂center＂)
        treeview.column(＂#3＂,width=100,anchor=＂center＂)

        treeview.heading(＂#0＂,text=＂no＂,anchor=＂center＂)
        treeview.heading(＂#1＂,text=＂name＂,anchor=＂center＂)
        treeview.heading(＂#2＂,text=＂birthday＂,anchor=＂center＂)
        treeview.heading(＂#3＂,text=＂tel＂,anchor=＂center＂)
        
        #검색한 정보를 표에 삽입
        for i in range(len(rows)):
            treeview.insert("","end",text=i+1,values=rows[i],iid=str(i)+"번")

#수정 함수
def update():
    global treeview #전역변수
    name=e_name.get()
    birthday=e_birthday.get()
    tel=e_tel.get()
    
    if(name!=＂＂ and birthday==＂＂): #이름으로 전화번호 변경
        Update=＂Update student set tel=＇%s＇ where name=＇%s＇＂ %(tel,name) #sql문
        cursor=con.cursor()
        cursor.execute(Update)
        con.commit()
        MessageBox.showinfo("Update Status","Updated Successfully") #성공을 알리는 알림 창

        treeview.destroy() #전체 표 삭제 후 다시 생성
        treeview=tkinter.ttk.Treeview(root,columns=("one,two,three","four"))
        treeview.pack()
        treeview["columns"]=("one","two","three","four")
        treeview.column("#0",width=100,anchor="center")
        treeview.column("#1",width=100,anchor="center")
        treeview.column("#2",width=100,anchor="center")
        treeview.column("#3",width=100,anchor="center")
        treeview.heading("#0",text="no",anchor="center")
        treeview.heading("#1",text="name",anchor="center")
        treeview.heading("#2",text="birthday",anchor="center")
        treeview.heading("#3",text="tel",anchor="center")

        ShowRecord() #테이블 조회
    
    elif(name==＂＂ and birthday!=＂＂): #생일으로 전화번호 변경
        Update=＂Update student set tel=＇%s＇ where birthday=＇%s＇＂ %(tel,birthday) 
        cursor=con.cursor()
        cursor.execute(Update)
        con.commit()
        MessageBox.showinfo(＂Update Status＂,＂Updated Successfully＂)
        
        treeview.destroy()
        treeview=tkinter.ttk.Treeview(root,columns=(＂one,two,three＂,＂four＂))
        treeview.pack()
        treeview[＂columns＂]=(＂one＂,＂two＂,＂three＂,＂four＂)
        treeview.column(＂#0＂,width=100,anchor=＂center＂)
        treeview.column(＂#1＂,width=100,anchor=＂center＂)
        treeview.column(＂#2＂,width=100,anchor=＂center＂)
        treeview.column(＂#3＂,width=100,anchor=＂center＂)

        treeview.heading(＂#0＂,text=＂no＂,anchor=＂center＂)
        treeview.heading(＂#1＂,text=＂name＂,anchor=＂center＂)
        treeview.heading(＂#2＂,text=＂birthday＂,anchor=＂center＂)
        treeview.heading(＂#3＂,text=＂tel＂,anchor=＂center＂)
        
        ShowRecord() #테이블 조회
 else: #수정 조건을 만족하지 않는 경우
        MessageBox.showinfo("Update Status","Some fields left blank") 
        
    e_name.delete(0,'end')
    e_birthday.delete(0,'end')
    e_tel.delete(0,'end’) 


#삭제 함수

def delete():
    global treeview
    name=e_name.get()
    if(e_name.get()==""): #이름을 입력하지 않은 경우
        MessageBox.showinfo(＂Delete Status＂,＂name is compolsary for delete＂) #삭제 실패를 알리는 알림 창
    else:        
        Delete=＂Delete from student where name=＇%s＇＂ %(name) #sql문
        cursor=con.cursor()
        cursor.execute(Delete)
        con.commit()
        MessageBox.showinfo(＂Delete Status＂,＂Deleted Successfully”) #삭제 성공을 알리는 알림 창
        
        treeview.destroy() #전체 표를 삭제하고 다시 생성
        treeview=tkinter.ttk.Treeview(root,columns=("one,two,three","four"))
        treeview.pack()
        treeview["columns"]=("one","two","three","four")
        treeview.pack()
        treeview["columns"]=("one","two","three","four")
        treeview.column("#0",width=100,anchor="center")
        treeview.column("#1",width=100,anchor="center")
        treeview.column("#2",width=100,anchor="center")
        treeview.column("#3",width=100,anchor="center")

        treeview.heading("#0",text="no",anchor="center")
        treeview.heading("#1",text="name",anchor="center")
        treeview.heading("#2",text="birthday",anchor="center")
        treeview.heading("#3",text="tel",anchor="center")
        
        ShowRecord() #전체 테이블 조회
        
        e_name.delete(0,'end')
        e_birthday.delete(0,'end')
        e_tel.delete(0,'end')


#버튼 만들기

insert=Button(root, text="Insert",font=("italic",10),bg="white",command=insert) #command=버튼을 눌렀을 때 실행될 함수
Insert.place(x=20,y=260) #버튼 위치 지정

Show=Button(root, text=＂Show＂,font=(＂italic＂,10),bg=＂white＂,command=ShowRecord)
show.place(x=80,y=260)

get=Button(root, text="Get", font=("italic",10),bg="white",command=get)
get.place(x=140,y=260)

update=Button(root, text="Update",font=("italic",10),bg="white",command=update)
update.place(x=190,y=260)

delete=Button(root, text="Delete",font=("italic",10),bg="white",command=delete)
delete.place(x=260,y=260)


#무한 루프
root.mainloop() #root를 무한루프 시킴
```
